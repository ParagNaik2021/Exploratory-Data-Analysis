# **Exploratory Data Analysis For Heart Disease**

The data science lifecycle for a project consist of seven steps which are problem definition, data collection, data preparation, data exploration, data modeling, model evaluation and model deployment.But among them Exploratory Data analysis is the most important one.

In this notebook, I will be doing extensive exploratory data analysis which is important for feature selection for building a Heart disease classifier model. To build the heart disease classifier by using UCI heart disease) dataset.

**Dataset Information**

<b>The dataset has 14 attributes:</b>

1.<b>Age</b>: age in years.<br>
2.<b>Sex</b>: sex (1 = male; 0 = female).<br>
3.<b>CP</b>: chest pain type (Value 0: typical angina; Value 1: atypical angina; Value 2: non-anginal pain; Value 3: asymptomatic).<br>
4.<b>Trestbps</b>: resting blood pressure in mm Hg on admission to the hospital.<br>
5.<b>Chol</b>: serum cholestoral in mg/dl.<br>
6.<b>Fbs</b>: fasting blood sugar > 120 mg/dl (1 = true; 0 = false).<br>
7.<b>Restecg</b>: resting electrocardiographic results (Value 0: normal; Value 1: having ST-T wave abnormality; Value 2: probable or definite left ventricular hypertrophy).<br>
8.<b>Thalach</b>: maximum heart rate achieved.<br>
9.<b>Exang</b>: exercise induced angina (1 = yes; 0 = no).<br>
10.<b>Oldpeak</b>: ST depression induced by exercise relative to rest.<br>
11.<b>Slope</b>: the slope of the peak exercise ST segment (Value 0: upsloping; Value 1: flat; Value 2: downsloping).<br>
12.<b>Ca</b>: number of major vessels (0-3) colored by flourosopy.<br>
13.<b>Thal</b>: thalassemia (3 = normal; 6 = fixed defect; 7 = reversable defect).<br>
14.<b>Target</b>: heart disease (1 = no, 2 = yes)
