# Attributes Correlation 

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True,linewidths=0.4)
plt.title("Attributes Correlation \n",fontsize=20)
plt.show()
```

# Level Nitrogen content in soil suitable for specific crops

```python
plt.figure(figsize=(20,8))
sns.set(rc={"axes.facecolor":"#283747","axes.grid":False})
sns.barplot(x='crop',y='N',data=df)
plt.title("Types of crop and suitable value of Nitrogen content in soil \n",fontsize=20)
plt.xticks(rotation=25)
plt.xlabel("Types of crops",fontsize=20)
plt.ylabel(" ratio of Nitrogen content in soil",fontsize=20)
plt.show()
```

# Catergorising crops based on fruit crops,pulses,cash crops,food crops

```python
df_fruits=df[(df['crop']=='mango') | (df['crop']=='apple') |(df['crop']=='banana') | (df['crop']=='pomegranate') |(df['crop']=='orange')|
  (df['crop']=='papaya')|(df['crop']=='watermelon')|(df['crop']=='grapes')|(df['crop']=='muskmelon')]
```

```python
df_pulses=df[(df['crop']=='mungbean')|(df['crop']=='chickpea')|(df['crop']=='lentil')|(df['crop']=='blackgram')|
            (df['crop']=='mothbeans')|(df['crop']=='kidneybeans')|(df['crop']=='pigeonpeas')]
```

```python
df_cash_crops=df[(df['crop']=='cotton')|(df['crop']=='coffee')|(df['crop']=='jute')|(df['crop']=='coconut')]
```

```python
df_food_crop=df[(df['crop']=='rice')|(df['crop']=='maize')]
```

```python
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in df_food_crop:
    if plotnumber<=7 :     # as there are 9 columns in the data
        ax = plt.subplot(4,3,plotnumber)
        sns.stripplot(x='crop',y=column ,data=df_food_crop)
        plt.xlabel(column,fontsize=20)
        plt.ylabel(column,fontsize=20)
        plt.xticks(np.arange(2),('rice','maize'),fontsize=18)
    plotnumber+=1
plt.show()
```

```python
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in df_cash_crops:
    if plotnumber<=7 :     # as there are 9 columns in the data
        ax = plt.subplot(4,3,plotnumber)
        sns.stripplot(x='crop',y=column ,data=df_cash_crops)
        plt.xlabel(column,fontsize=20)
        plt.ylabel(column,fontsize=20)
        plt.xticks(np.arange(4),('coconut','cotton','jute','coffee'),fontsize=18)
    plotnumber+=1
plt.show()
```

```python
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in df_fruits:
    if plotnumber<=7 :     # as there are 9 columns in the data
        ax = plt.subplot(5,2,plotnumber)
        sns.stripplot(x='crop',y=column ,data=df_fruits)
        plt.xlabel(column,fontsize=20)
        plt.ylabel(column,fontsize=20)
        plt.xticks(np.arange(9),('pomogranate','banana','mango','grapes','watermelon',' muskmelon','apple','orange','papaya'))
    plotnumber+=1
plt.show()
```

```python
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in df_pulses:
    if plotnumber<=7 :     # as there are 9 columns in the data
        ax = plt.subplot(5,2,plotnumber)
        sns.stripplot(x='crop',y=column ,data=df_pulses)
        plt.xlabel(column,fontsize=20)
        plt.ylabel(column,fontsize=20)
        plt.xticks(np.arange(7),('chickpea','kidneybean','pigeonbean','mothbean','mungbean','blackgram','lentil'),fontsize=12)
    plotnumber+=1
plt.show()
```
